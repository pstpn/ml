```python
import pandas as pd

df = pd.read_csv("data/Chemical Composion of Ceramic.csv")
display(df)
df.info()
```

#### Drop useless columns ("Ceramic Name" - unique, "Part" - binary feature)

```python
df.drop(labels=["Ceramic Name", "Part"], axis=1, inplace=True)
```

#### KMeans++ clustering with silhouette metric

```python
from kmeanspp import KMeansPP
from metrics import silhouette


max_iter, max_clusters = 100, 10
for count in range(2, max_clusters + 1):
    clustered_df = df.copy()
    _, ok = KMeansPP.clustering(clustered_df, count, max_iter, random_state=42)
    if not ok:
        print(f"Algorithm did not converge within the maximum number of iterations: {count} clusters")
    else:
        score = silhouette(clustered_df, df.columns.to_list())["Silhouette"]
        print(f"Silhouette score for {count} clusters: {score.mean()}")
```
